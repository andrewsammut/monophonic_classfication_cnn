```python
from os import listdir
from os.path import isdir, join
import wave
import contextlib
import librosa
import soundfile as sf

# Dataset path and view possible targets
dataset_path = 'C:\\Users\\Greg\\ML\\Technical\\project\\git2\\monophonic_classfication_cnn\\padded'
num_samples = 0
durations = []
sampdurations = []


for inst in listdir(dataset_path):
    #inst2 = inst + "\\" + inst
    pitch = join(dataset_path, inst)
    print(inst)
    for target in listdir(pitch):
        fname = join(pitch, target)
        #print(fname)
        with contextlib.closing(wave.open(fname,'r')) as f:
            frames = f.getnframes()
            rate = f.getframerate()
            duration = frames / float(rate)
            #print(rate)
            sampdurations.append(duration)
    maxd = max(sampdurations)
    mind = min(sampdurations)
    avg = sum(sampdurations) / len(sampdurations)
    print("max:", maxd)
    print("min:", mind)
    print("average:", avg)
    durations.extend(sampdurations)
    sampdurations.clear()
    print()
print("TOTAL Max / min values")
print("max max:", max(durations))
print("Min min:", min(durations))
sumd = sum(durations)
leng = len(durations)
print("average:", sumd / leng)
#for target in listdir(dataset_path):
#    print(len(listdir(join(dataset_path, target))))
#    num_samples += len(listdir(join(dataset_path, target)))
#print('Total samples:', num_samples)
```
